1. Analisis Jaringan Sosial:  
Teori graf dapat digunakan untuk menganalisis jaringan sosial, di mana individu diwakili sebagai node, dan hubungan mereka diwakili sebagai edge. Ini dapat membantu mengidentifikasi cluster, influencer, dan tren dalam jaringan.

```python
import networkx as nx
import matplotlib.pyplot as plt

def calculate_influencer(G):
    """
    Menghitung influencer dalam jaringan sosial berdasarkan derajat sentralitas.
    """
    centralities = nx.degree_centrality(G)
    influencer = max(centralities, key=centralities.get)
    return influencer

G = nx.Graph()

# Tambah node dan edge
G.add_node("Bima")
G.add_node("Zahra")
G.add_node("Ikhsan")
G.add_node("Niko")
G.add_edge("Bima", "Zahra")
G.add_edge("Zahra", "Ikhsan")
G.add_edge("Bima", "Ikhsan")
G.add_edge("Zahra", "Niko")

# Visualisasi graf
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_color='black', linewidths=1, font_size=12)

# Hitung koefisien clustering
print("Koefisien clustering Ikhsan =", nx.clustering(G, "Ikhsan"))  # Output: 1.0

# Hitung panjang jalur terpendek antara dua node
print("Panjang jalur terpendek antara Bima dan Niko =", nx.shortest_path_length(G, "Bima", "Niko"))  # Output: 2

# Mencari influencer
influencer = calculate_influencer(G)
print("Influencer =", influencer) # Output: Zahra

plt.show()
```

2. Optimasi Jaringan Lalu Lintas:  
Teori graf dapat digunakan untuk mengoptimalkan jaringan lalu lintas, di mana persimpangan diwakili sebagai node, dan jalan diwakili sebagai edge. Ini dapat membantu mengurangi kemacetan.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Tambah node dan edge
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_edge("A", "B", capacity=5)
G.add_edge("B", "C", capacity=10)
G.add_edge("A", "C", capacity=15)

# Visualisasi graf
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, edge_color='black', linewidths=1, font_size=12)
labels = nx.get_edge_attributes(G, 'capacity')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Hitung aliran maksimum
flow_value, flow_dict = nx.maximum_flow(G, "A", "C")
print("Aliran maksimum dari A ke C =", flow_value)  # Output: 20

plt.show()
```
